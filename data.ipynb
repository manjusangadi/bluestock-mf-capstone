```python
import pandas as pd

fund_master = pd.read_csv(
    "data/raw/01_fund_master.csv"
)

nav_history = pd.read_csv(
    "data/raw/02_nav_history.csv"
)
```

```python
# 2. Check Column Names

# First check the exact column names.

print(fund_master.columns)

print(nav_history.columns)
```

```python
# 3. Create Sets of AMFI Codes
master_codes = set(
    fund_master["amfi_code"]
)

nav_codes = set(
    nav_history["amfi_code"]
)
print(master_codes)
print(nav_codes)
```

```python
missing_code = (master_codes - nav_codes)
print(missing_code)
# if result shows set() means All AMFI codes exist in nav_history.
```

```python
# 6. Count Total Codes
print(
    "Fund Master Codes:",
    len(master_codes)
)

print(
    "NAV History Codes:",
    len(nav_codes)
)
```
